- 這裏是一個建立`正弦波`圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸的數據點
x = np.linspace(0, 2*np.pi, 100)

# 計算正弦值
y = np.sin(x)

# 繪製圖表
plt.plot(x, y)
plt.xlabel("x")  # 設定 x 軸標籤
plt.ylabel("sin(x)")  # 設定 y 軸標籤
plt.title("sin plot")  # 設定圖表標題
plt.grid(True)  # 顯示格線
plt.show()  # 顯示圖表
```
